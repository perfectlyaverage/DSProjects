## Notes:
### We need to have same target value for same identifier of household
### Unfortunately, for some families, target are not the same for the household (even with the same idhogar)
### Thus, we need to assign the target to all the member of the idhogar
### Using the head of the house's target variable

```python

# Calculating the idhogar key per family... should be one unique id
target_equal = train_df.groupby('idhogar')['Target'].apply(lambda x: x.nunique() == 1)

# Calculating the familes that have multiple unique ids
target_not_equal = target_equal[target_equal == False]

# The total amnt of mismatched values
print(len(target_not_equal))
target_not_equal

```

# Fixing Values

```python
#Seting poverty level of the members and the head of the house within a family.
for household in target_not_equal.index:
    true_target = int(train_df[(train_df['idhogar'] == household) & (train_df['parentesco1'] == 1)]['Target'])
    train_df.loc[train_df['idhogar'] == household, 'Target'] = true_target
```

```python
# Selecting input and target data for training
input_data=train_df.iloc[:,0:-1]
target_data=train_df.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,classification_report

#Splitting by train:test = 80:20
x_train,x_test,y_train,y_test=train_test_split(input_data,target_data,test_size=0.2,random_state=1)
#using 100 random trees.
rmclassifier = RandomForestClassifier(n_estimators=100)
```

```python
#Fitting the model
rmclassifier.fit(x_train,y_train)
```

```python
#Storing predicted Values
y_predict = rmclassifier.predict(x_test)
```

```python
#accuracy using RandomForestClassifier
percentage = accuracy_score(y_test,y_predict)*100
percentage.round(2)
print(f'The accuracy of the model is :{percentage.round(2)}%')
```

```python
#Applying Kfold here with 10 splitts and on 100 trees
from sklearn.model_selection import KFold,cross_val_score

kfold=KFold(n_splits=10,random_state=1,shuffle=True)

rmclassifier=RandomForestClassifier(n_estimators=100,random_state=1)
list_of_results = cross_val_score(rmclassifier,input_data,target_data,cv=kfold,scoring='accuracy')
print(f'Ten Fold Result List:\n{list_of_results}')
```

```python
perct = list_of_results.mean()*100
print(f'The model accuracy after KFold is : {perct.round(2)}%')
```
